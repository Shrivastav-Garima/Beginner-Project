```python
import matplotlib.pyplot as plt
```

```python
data['transaction_date'] = pd.to_datetime(data['transaction_date'])  
monthly_data = data.groupby(data['transaction_date'].dt.to_period('M')).sum(numeric_only=True)
monthly_data['transaction_qty'].plot(kind='line')  
plt.show()
```

```python

product_sales = data.groupby('product_type').sum(numeric_only=True)['transaction_qty']
```

```python
top_products = product_sales.sort_values(ascending=False)
```

```python
top_10_products = top_products.head(10)
print(top_10_products)
```

```python
plt.figure(figsize=(10, 6))
top_10_products.plot(kind='bar', color='orange')
plt.title("Top 10 Selling Products")
plt.xlabel("Product Type")
plt.ylabel("Total Transaction Quantity")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_10_products.plot(kind='barh', color='lightgreen')
plt.title("Top 10 Selling Products (Horizontal View)")
plt.xlabel("Total Transaction Quantity")
plt.ylabel("Product Type")
plt.show()
```
